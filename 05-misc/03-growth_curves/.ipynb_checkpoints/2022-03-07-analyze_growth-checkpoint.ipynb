```python
layout_df = import_dataframes_from_excel_file('eLW008_ cultures.xlsx')['Sheet1']
first_col = layout_df.columns[0]
blocks = None
for unique in layout_df.loc[:,first_col].unique():
    query = layout_df[ layout_df[first_col] == unique ]
    query = pd.melt( query.iloc[:,:], id_vars=[first_col,'/'])
    query.columns = ['field','row','col',unique]
    query = query.query("row != '/'").reset_index(drop=True).drop('field',axis=1).dropna()
    blocks = pd.concat([blocks,query],axis=1)
blocks_df = blocks.loc[:,~blocks.columns.duplicated()]
blocks_df['col'] = blocks_df['col'].apply(lambda x: str(int(x)).zfill(2))
blocks_df['well'] = blocks_df['row'] + blocks_df['col']
blocks_df.drop('row',axis=1,inplace=True)
blocks_df.head()
```

```python
start_lines = []
end_lines = []
prev_end = False
with open('2022-03-04-growth_curves.txt',encoding='windows-1252') as file:
    for line_number, line in enumerate(file, 1):
        if '~End' in line: # file lines that have '~End' specify where the data starts in the following line
            end_lines.append(line_number)
            prev_end = True
        if prev_end == True and 'Plate' in line: # where to start data
            start_lines.append(line_number)
            prev_end = False
            
blocks = []
for start_line in start_lines:
    for end_line in end_lines:
        if start_line < end_line:
            blocks.append({'start':start_line,'end':end_line})
            break
```

```python
COLUMNS = 8
ROWS = 12
columns = ascii_uppercase[:COLUMNS]
rows = range(1,ROWS + 1)
wells = [column + str(row) for column in columns for row in rows]
well_dtypes = {well:str for well in wells}
well_dtypes.update({'Time(hh:mm:ss)':str,'Temperature(¡C)':str})
df = pd.read_csv(
    filepath_or_buffer='2022-03-04-growth_curves.txt',
    sep='\t',
    encoding='windows-1252',
    skiprows=11445,
    dtype=well_dtypes,
    # na_values=0
).iloc[:22849,:-1]


wells_zeros = [column + str(row).zfill(2) for column in columns for row in rows]
wells_map = {well:wells_zero for well in wells for wells_zero in wells_zeros if well[0] + well[1:].zfill(2) == wells_zero}
df.rename(columns=wells_map,inplace=True)
df.rename(columns={'Temperature(¡C)':'Temperature','Time(hh:mm:ss)':'Time'},inplace=True)
df.dropna(inplace=True)

def time_to_minutes(time):
    fields = len(time.split(':'))
    seconds = int(time.split(':')[-1]) / 60
    minutes = int(time.split(':')[-2])
    
    if fields == 2:
        return minutes + seconds
    if fields == 3:
        hours = int(time.split(':')[0]) * 60
        return hours + minutes + seconds
    else:
        raise Exception('error encountered while formatting time')
        
df['minutes'] = df['Time'].apply(lambda x: time_to_minutes(x))
df.drop(['Time','Temperature'],axis=1,inplace=True)
df = pd.melt(df,id_vars=['minutes'],value_vars=wells_zeros,var_name='well',value_name='OD')
df['OD600'] = df['OD'].astype(float)
merged_df = blocks_df.merge(df,on='well')
```

```python
biovar_map = {"sLW002":"BV1",
              "sLW005":"BV2",
              "sLW006":"BV3",
             "BLANK":"BLANK"}
merged_df['biovar'] = merged_df['strain'].apply(lambda x: biovar_map[x])

carbon_source_map = {'MOPS+glucose':'MOPS+glucose',
                     'MOPS+succinate':'MOPS+succinate',
                     'PITA':'MGYS',
                     'LB':'LB',
                     'R2A':'R2A'}
merged_df['growth_media'] = merged_df['condition'].apply(lambda x: carbon_source_map[x])
```

```python
merged_df['hours'] = merged_df['minutes'].apply(lambda x: x/60.0)
```

```python
g = sns.relplot(kind='line',data=merged_df[merged_df['minutes'] % 30 == 0].query("hours < 60"),col='biovar',x='hours',y='OD600',row='growth_media',height=2.5,aspect=1.4)
g.savefig("2022-03-07-eLW008-growth_curves.png",dpi=300)
```
