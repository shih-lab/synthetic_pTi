```python
dfs = import_dataframes_from_excel_file('eLW008_ MIC measurements.xlsx')
df = dfs['layout'].merge(dfs['measurement'],on=['row','column'])
df['strain-replicate'] = df['strain'] + '-' + df['replicate'].astype(str)
df['log_abx_conc'] = np.log(df['antibiotic_concentration-mg/L'])
blank_OD = dfs['blank_well']['OD'].item()
df['rel_OD'] = df['OD'].apply(lambda x: abs(x - blank_OD))
df.drop(105,inplace=True)
cmap = sns.cubehelix_palette(start=2.6, rot=.1,gamma=1.0,hue=0.8,light=0.8,dark=0.2,n_colors=12)

biovar_map = {"sLW002":"BV1",
              "sLW005":"BV2",
              "sLW006":"BV3"}
df['biovar'] = df['strain'].apply(lambda x: biovar_map[x])
```

```python
grid = sns.FacetGrid(data=df,col='antibiotic_name',row='biovar',sharex='col',sharey='row',height=3.5,aspect=1.2)
grid.map_dataframe(sns.barplot,x='antibiotic_concentration-mg/L',y='rel_OD',palette=cmap)
grid.set_xticklabels(rotation=90)
grid.set_ylabels('Relative OD600')
grid.set_xlabels('Antibiotic Concentration (ug/mL)')
grid.savefig('2022-03-15-eLW008-MIC_measurements.png',dpi=300)
plt.show()
```
